```python
# Water Jug Problem using DFS

# Capacities of jugs
CAP_A = 4
CAP_B = 3

# Target amount
TARGET = 2

# DFS function
def dfs(a, b, visited, path):
    
    if (a, b) in visited:
        return False
    
    visited.add((a, b))
    
    path.append((a, b))
    print(f"Current State: Jug A = {a}, Jug B = {b}")
    if a == TARGET or b == TARGET:
        print("\nGoal reached!")
        print("Path:")
        for state in path:
            print(state)
        return True
    
    print("Rule: Fill Jug A")
    if dfs(CAP_A, b, visited, path.copy()):
        return True

    print("Rule: Fill Jug B")
    if dfs(a, CAP_B, visited, path.copy()):
        return True
    print("Rule: Empty Jug A")
    if dfs(0, b, visited, path.copy()):
        return True

    print("Rule: Empty Jug B")
    if dfs(a, 0, visited, path.copy()):
        return True
    
    print("Rule: Pour Jug A → Jug B")
    transfer = min(a, CAP_B - b)
    if dfs(a - transfer, b + transfer, visited, path.copy()):
        return True
    print("Rule: Pour Jug B → Jug A")
    transfer = min(b, CAP_A - a)
    if dfs(a + transfer, b - transfer, visited, path.copy()):
        return True
    
    return False

visited = set()
path = []

print("Water Jug Problem using DFS\n")

dfs(0, 0, visited, path)
```
