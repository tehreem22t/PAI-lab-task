```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# Load image
img = Image.open(r"C:\Users\Home\Downloads\bird.jpg").convert("L")
gray_np = np.array(img)

# Compute gradients
gx = np.abs(np.diff(gray_np, axis=1))
gy = np.abs(np.diff(gray_np, axis=0))

# Make both same shape
min_rows = min(gx.shape[0], gy.shape[0])
min_cols = min(gx.shape[1], gy.shape[1])

gx = gx[:min_rows, :min_cols]
gy = gy[:min_rows, :min_cols]

# Edge magnitude
edges = np.hypot(gx, gy)

# Normalize
edges = (edges / edges.max() * 255).astype(np.uint8)

# Display
plt.imshow(edges, cmap='gray')
plt.axis('off')
plt.title("Edge Detection")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
img = Image.open(r"C:\Users\Home\Downloads\bird.jpg")
img_np = np.array(img)
cropped = img_np[50:250, 100:300]

plt.figure(figsize=(8,4))

plt.subplot(1,2,1)
plt.imshow(img)
plt.title("Original Image")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(cropped)
plt.title("Cropped Image")
plt.axis("off")

plt.show()
```
